* C언어는 코드 블록을 표시하기 위해 중괄호 {}를 사용하지만 파이썬은 코드 블록을 들여쓰기로 처리한다.

```python
# 비만도(BMI) 계산에서 코드가 아족에 속할수록 들여쓰기가 커진다.
def cal_bmi(height, weight):
  bmi = weight / (height * height)
  print("BMI:{:.2f}".format(bmi),end="")
  if bmi < 18.5:
      print("마른체형")
  elif 18.5<=bmi<25.0:
      print("표준")
  elif 25.0 <= bmi < 30.0:
      print("비만")
  else:
      print("고도비만")

height = input("키 (m): ")
weight = input("체중 (kg): ")
cal_bmi(float(height), float(weight))       
```

```python
s = 'Hello'
s1 = s.rjust(20) # 전체자리수 20
print(s1)
s1 = s.rjust(20,'*') # 우측정렬하고 빈칸을 *로 채움
print(s1)
s1 = s.ljust(20,'-') # 좌측정렬하고 빈칸은 -로 채움
print(s1)
s1 = s.center(20,'=') # 중앙정렬하고 빈칸을 =로 채움
print(s1)
```



* sort(key = None, reverse=False)는 리스트의 항목을 정렬한다. 인수 reverse는 정렬 순서를 지정하며 그 값이 True이면 내림차순으로 정렬하며, 인수가 없으면 오름차순으로 정렬한다.
알파벳 문자의 정렬 순서는 문자의 아스키 코드로 결정된다.
아스키 코드는 A가 65, 문자마다 1씩 증가해 Z가 90, a가 97, z가 122이다.

스택으로 사용



* 스택은 후입선출(LIFO : Last In dirst Out) 방식으로 동작해서 스택에 마지막에 들어온 항목이 먼저 나간다. 리스트를 스택으로 사용하는 것은 쉽다. append()를 사용해 스택 맨 위에 항목을 추가하고 pop()을 사용해 스택의 맨 위에서 항목을 꺼낸다.



```python
stack = [1, 2, 3]
stack.append(4)
stack.append(5)
print(stack)
print(stack.pop()) # 마지막에 추가한 항목 추출
print(stack)
```

큐로 사용



* 큐는 선업선출(FIFO: Forst In First Out) 방식으로 동작해서 큐에 먼저 들어간 항목이 먼저 나온다. 리스트를 큐로 사용하는 것이 가능은 하지만 효율적이지 못하다. 리스트의 끝에 추가하고(append) 빼내는 것(pop)은 처리속도가 빠르지만 리스트의 처음에 삽입하고(insert) 빼내려면(pop) 리스트의 모든 항목이 한 번 이동해야 돼 처리속도가 느리다.





* 모듈 collections, deque는 이를 빨리 처리한다.



```python
from collections import deque

q = deque([1, 2, 3, 4, 5]) # 리스트를 큐로 변환
print(q)

q.append(6)
q.append(7)
print(q)

print(q.popleft()) # 맨 처음 항목을 추출
print(q)
print(q.pop()) # 맨 마지막에 추가된 항목을 추출
print(q)
```

**집합**

* 집합은 여려 개의 원소를 저장하는데 중복과 순서가 없다.
* 중복을 허용하지 않기 때문에 중복된 데이터를 넣어도 하나만 저장된다.
* 파이썬은 set과 frozenset의 두 가지 집합 자료형을 제공한다.
* set은 변경가능하고 frozenset는 변경불가이
* set는 set() 함수나 {}를 이용해 만들며 빈 집합을 만들 수 있다.

```python
a = set() # 빈 집합 생성 
print(a)

b = {1, 2, 3}
print(b)

c = b.copy() # b 집합을 그대로 복제
print(c)
```

set() 함수를 사용해 문자열을 집합으로 변환할 수 있으며, 변환 결과에 순서와 중복이 없다.

* 리스트를 집합으로 변환할 수 있지만(리스트의 모든 항목이 집합의 원소로 변환) 리스트가 집합의 원소가 될 수 없다.


```python
a = [1, 2, 3]
b = [4, 5]

A = (6, 7, 8)
B = (9, 10)
c = set(a) # 리스트를 집합으로 변환가능
print(c)

d = set(A)
print(d)

e = {A, B} # 튜플은 변경불가 자료형으로 집합의 원소가 될 수 있음
print(e)

f = {a, b} # 리스트는 변경가능 자료형으로 집합의 원소가 될 수 없음
```

* 주어진 객체(리스트, 집합, 튜플)에 대해 합집합 연산을 하려면 update()를 사용한다.
* 집합의 원솔 개수를 알려면 len()을 사용한다.

```python
a = {1, 2}
a.add(3)
print(a)

a.update([4, 5]) # 리스트와 합집합
print(a)

a.update({0, 6}) # 집합과 합집합
print(a)

a.update((6, 7)) # 튜플과 합집합
print(a, len(a))
```

* 집합의 원소를 모두 제거하려면 clear()를 사용한다.
* 원소 한 개를 제거하려면 discard() 또는 remove()를 사용한다.
* discard는 제거할 원소가 없을 때 무시하고 넘어가지만 remove()는 오류를 발생한다.
* 원소 한 개를 제거하고 이를 반환 받으려면 pop()을 사용한다.
* pop()은 집합의 가종 왼쪽 원소부터 꺼낸다.

```python
a = {1, 2}
a.clear()
print(a)

a = {1, 2, 3, 4}

a.remove(1) #a에서 1 제거
print(a)

a.discard(1) # a에 제거할 1이 없지만 넘어감
print(a)

b = a.pop()
print(a)
print(b)
```

**집합연산**


* 합집한 : union, |
* 교집합 : intersection, &
* 차집합 : diffrence, -
* 대칭차집합 : symmetric_diffrence, ^

```python
a = {1, 3, 5}
b = {5, 6}
print(a.union(b)) # 합집합
print(a|b) # 합집합

print(a.intersection(b)) # 교집합
print(a & b) # 교집합

print(a.difference(b)) # 차집합
print(a - b) # 차집합

print(a.symmetric_difference(b)) # 대칭 차집합
print(a ^ b) # 대칭 차집합
```

부분 집합과 공집합 여부에 대해 평가할 수 있다.

```python
a = {1, 3, 5}
b = {5}
print(a.issuperset(b)) # A가 B의 상위(super) 집합(set) 
print(b.issubset(a)) # B가 A의 하위(sub) 집합(set)
print(b.isdisjoint(a)) # 공집합인지를 평가
```

return 문장이 여러 개 있을 수 있고 return 문장이 실행되면 함수 실행은 종료된다.

```python
def cal(s, a, b) :
  if s == '+':
    return(a+b)
  elif s == '-':
    return(a-b)
  elif s == '*':
    return(a*b)
  elif s == '/':
    return(a/b)

print(cal('+', 5, 2), cal('-', 5, 2), cal('*', 5, 2), cal('/', 5, 2))
```

함수에서 전역변수 사용

```python
total = 10 # 전역변수로 어디에서도 참조 가능

def sum(a, b) :
  result = 0 # 지역변수로 sum() 내부에서만 유효
  result = a + b + total
  return(result)

total = sum(1,2)
print(total)
```

```python
total = 0

def sum(a, b) :
  # 함수 내에서 전역변수 total에 값을 저장하려면 global total라는 선언 필요
  global total
  total = a + b + total
  return

sum(1, 2)
print(total)
```

함수 내부에 global total을 선언하지 않으면 total 지역변수로 인식해서 오류가 발생한다.

```python
total = 0

def sum(a, b) :
  total = a + b + total
  return

sum(1, 2)
print(total)
```

지도학습 : 선형 회로

1.1 데이터 준비

1.1.1 데이터 생성

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# 데이터 샘플 수 지정
SAMPLES = 300

# 시드 값을 지정, 동일한 랜덤 패턴을 구현
SEED = 1
np.random.seed(SEED)

# 사인 값 발생을 위한 0 ~ 2pi 범위의 x 값을 균일 분포 난수로 생성
x = np.random.uniform(low=0, high=2*math.pi, size=SAMPLES).astype(np.float32)

# 사인 값을 계산
y = x + 1.5*np.sin(x).astype(np.float32)

#데이터를 그래프로 표현. 인수 'b.'는 파란색 도트로 그리도록 함
plt.plot(x, y, 'b.')
plt.show()
```

```python
# y 값에 무작위로 작은 값을 더함
y += 1.*np.random.random(SAMPLES) # 가우시안 정규 분포

# 그래프로 표현 
plt.plot(x, y, 'b.')
plt.show()
```

1.1.2 데이터 분할

```python
# 분할 지점 지정
TRAIN_SPLIT = int(0.6 * SAMPLES) # 검증 데이터 시작 위치 지정
TEST_SPLIT  = int(0.2 * SAMPLES + TRAIN_SPLIT) # 테스트 데이터 시작 위치 지정

# 데이터 분할을 위해 np.split()을 사용
# np.split()의 둘째 인수는 데이터가 분할될 인덱스를 배열로 표현
# 두개의 인덱스를 기준으로 데이터를 3개로 분할
x_train, x_validate, x_test = np.split(x, [TRAIN_SPLIT, TEST_SPLIT])
y_train, y_validate, y_test = np.split(y, [TRAIN_SPLIT, TEST_SPLIT])

#분할된 데이터 크기 출력
print(x_train.shape, x_validate.shape, x_test.shape)
print(y_train.shape, y_validate.shape, y_test.shape)

# 1D -> 2D 배열로 변환
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_validate = x_validate.reshape(-1,1)
y_validate = y_validate.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
print(x_train.shape, x_validate.shape, x_test.shape)
print(y_train.shape, y_validate.shape, y_test.shape)
```

1.2 선형 회귀 모델 구축

1.2.1 선형 회귀 모델의 학습

```python
from sklearn.linear_model import LinearRegression

# 선형 회귀 모델 구성
regr = LinearRegression()
print(regr)

# 훈련 데이터로 선형 회귀 모델을 학습함
regr.fit(x_train, y_train)

# y = regr.coaf_*x + regr.intercept_
print("학습된 계수:", regr.coef_, "학습된 절편:", regr.intercept_)

# 선형 회귀 예측
y_train_pred = regr.predict(x_train)
print("예측 값:", y_train_pred[0], "계산 값:", x_train[0]*regr.coef_ + regr.intercept_)
print(x_train[0], y_train[0], y_train_pred[0])
```

1.2.2 선형 회귀 모델의 평가

```python
from sklearn.metrics import mean_squared_error, r2_score

# 훈련 데이터에 대한 예측(추론)
y_train_pred = regr.predict(x_train)

# 검증 데이터에 대한 예측(추론)
y_validate_pred = regr.predict(x_validate)

# 시험 데이터에 대한 예측(추론)
y_test_pred = regr.predict(x_test)

# 평균제곱오차(MSE)
print('MSE: 훈련 %.2f, 검증 %.2f, 시험 %.2f ' % (mean_squared_error(y_train, y_train_pred),mean_squared_error(y_validate, y_validate_pred),mean_squared_error(y_test,y_test_pred)))

# 결정계수(R2)
print('R2 score: 훈련 %.2f, 검증 %.2f, 시험 %.2f ' %(r2_score(y_train, y_train_pred),r2_score(y_validate, y_validate_pred), r2_score(y_test, y_test_pred)))
```

```python
import numpy as np 
import matplotlib.pyplot as plt

z = np.array([1, 3, 5]) # 범주가 3개인 경우

def softmax(x):
  exp_x = np.exp(x)
  sum = np.sum(exp_x)
  y = exp_x / sum
  print("maxsft outputs:", y)
  print("sum of maxsoft outputs:", np.sum(y))
  return y
plt.title("Softmax Funtion")
plt.bar(z, softmax(z))
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# 데이터 샘플 수 지정
SAMPLES = 100

# 시드 값을 지정, 동일한 랜덤 패턴 구현
SEED = 1
np.random.seed(SEED)

# 0부터 3까지 범위에서 일정 간격으로 x 값을 생성
x = np.linspace(0, 3, SAMPLES)

# 생성된 x의 값이 순서를 따르지 않도록 값을 섞음
np.random.shuffle(x)

# y 값을 계산, 가우시안 정규 분포의 잡음 추가
y = 1.5*(1- np.exp(-4.5*x).astype(np.float32)) + 0.1*np.random.random(SAMPLES)

# 데이터를 그래프로 표현. 인수 'b.'는 파란색 도트로 그리도록 함
fig = plt.figure(figsize=(6, 4))
plt.scatter(x, y, s=4, alpha=1, color = 'blue')
plt.show()
```

5.2.3 분류 의사결정트리의 구축

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(criterion = 'entropy', max_depth=4, min_samples_leaf=2,
                                min_samples_split=2, random_state = 0)

dt_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

# 훈련 데이터의 대한 추론
y_train_pred = dt_clf.predict(x_train)

[[TN, FP], [FN, TP]] = confusion_matrix(y_train, y_train_pred)
TN, FP, FN, TP = confusion_matrix(y_train, y_train_pred).ravel()
print('---------Train---------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d}, False Negative(FN)={:3d} '.format(TP, FP, TN, FN))
accuracy = (TP + TN) / (TP+FP+TN+FN)
precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))

# 검증 데이터에 대한 추론 
y_validate_pred = dt_clf.predict(x_validate)
TN, FP, FN, TP = confusion_matrix(y_validate, y_validate_pred).ravel()

print('---------Validate---------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d}, False Negative(FN)={:3d}'.format(TP, FP, TN, FN))
accuracy = (TP + TN) / (TP+FP+TN+FN)
precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))

# 시험 데이터에 대한 추론
y_test_pred = dt_clf.predict(x_test)
TN, FP, FN, TP = confusion_matrix(y_test, y_test_pred).ravel()
print('---------Test---------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d}, False Negative(FN)={:3d}'.format(TP, FP, TN, FN))
accuracy = (TP + TN) / (TP+FP+TN+FN)
precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))
```

```python
from sklearn.metrics import confusion_matrix

# 훈련 데이터에 대한 추론
y_train_pred = rfc.predict(x_train)

[[TN, FP], [FN, TP]] = confusion_matrix(y_train, y_train_pred)
# TN, FP, FN, TP = confusion_matrix(y_train, y_train_pred).ravel()
print('------Train------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d}, False Negative(FN)={:3d} '.format(TP,FP,TN,FN))
accuracy = (TP+TN) / (TP+FP+TN+FN)
precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))

# 검증 데이터에 대한 추론
y_validate_pred = rfc.predict(x_validate)
TN, FP, FN, TP = confusion_matrix(y_validate, y_validate_pred).ravel()

print('------Validate------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d},False Negative(FN)={:3d} '.format(TP,FP,TN,FN))
accuracy = (TP+TN) / (TP+FP+TN+FN)
precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))

# 시험 데이터에 대한 추론
y_test_pred = rfc.predict(x_test)
TN, FP, FN, TP = confusion_matrix(y_test, y_test_pred).ravel()

print('------Test------')
print('True Positive(TP)={:3d}, False Positive(FP)={:3d}, True Negative(TN)={:3d},False Negative(FN)={:3d} '.format(TP,FP,TN,FN))
accuracy = (TP+TN) / (TP+FP+TN+FN)

precision = TP / (TP+FP)
recall = TP / (TP+FN)
print('Accuracy={:0.3f}, Precision={:0.3f}, Recall={:0.3f}'.format(accuracy, precision, recall))
```

```python
# 분할된 지점 지정
SAMPLES = x.shape[0]
print(SAMPLES)

TRAIN_SPLIT = int(0.6 * SAMPLES) # 검증 데이터 시작 위치 지정
TEST_SPLIT = int(0.2 * SAMPLES + TRAIN_SPLIT) # 테스트 데이터 시작 위치 지정

# 데이터 분할 np.split(x, [i,j]): x를 0~i~1, i~j~1, j~끝으로 분할
x_train, x_validate, x_test = np.split(x, [TRAIN_SPLIT, TEST_SPLIT]) # 원래 데이터
y_train, y_validate, y_test = np.split(y, [TRAIN_SPLIT, TEST_SPLIT])

# 분할된 데이터 크기 출력
print(x_train.shape, x_validate.shape, x_test.shape)
print(y_train.shape, y_validate.shape, y_test.shape)
```
